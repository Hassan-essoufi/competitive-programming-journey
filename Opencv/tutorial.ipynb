```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Reading images
image = cv2.imread('assets/image1.png')
cv2.imshow('reading image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Writing images
image = cv2.imread('assets/image1.png')
new_image = image + 12
cv2.imwrite('outputs/new_image.png', new_image)
```

```python
# Reading video
video = cv2.VideoCapture('assets/video.mp4')

while True:
    ret, frame = video.read()

    if not ret:
        break
    cv2.imshow('reading video', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```

```python
# Writing video 
video = cv2.VideoCapture('assets/video.mp4')

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
new_video = cv2.VideoWriter('outputs/new_video.mp4', fourcc, 30, (width, height))

while True:
    ret , frame = video.read()
    if not ret:
        break 
    frame = frame + 30
    new_video.write(frame)

new_video.release()
video.release()
cv2.destroyAllWindows()

```
