```python
# Reading video
video = cv2.VideoCapture('assets/video.mp4')

while True:
    ret, frame = video.read()

    if not ret:
        break
    cv2.imshow('reading video', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```

```python
# Writing video 
video = cv2.VideoCapture('assets/video.mp4')

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
new_video = cv2.VideoWriter('outputs/new_video.mp4', fourcc, 30, (width, height))

while True:
    ret , frame = video.read()
    if not ret:
        break 
    frame = frame + 30
    new_video.write(frame)

new_video.release()
video.release()
cv2.destroyAllWindows()
```

```python
# Colors space
image = cv2.imread('assets/image1.png')

# RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# HSV
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# lab
image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

plt.figure(figsize=(10,9))

plt.subplot(3,2,1)
plt.imshow(image)
plt.title("BGR")
plt.axis('off')

plt.subplot(3,2,2)
plt.imshow(image_rgb)
plt.title("RGB")
plt.axis('off')

plt.subplot(3,2,3)
plt.imshow(image_gray, cmap="gray")
plt.title("GrayScale")
plt.axis('off')

plt.subplot(3,2,4)
plt.imshow(image_hsv)
plt.title("HSV")
plt.axis('off')

plt.subplot(3,2,5)
plt.imshow(image_lab)
plt.title("LAB")
plt.axis('off')
```

```python
# Color thresholding
image =  cv2.imread("assets/image1.png")
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

upper = np.array([100, 100,255], dtype=np.uint8)
lower = np.array([140, 255, 255], dtype=np.uint8)

mask = cv2.inRange(image_hsv, lower, upper)
result = cv2.bitwise_and(image_hsv, image_hsv,mask=mask)
plt.imshow(result)
```

```python
# Flipping, Rotation and Cropping
image =cv2.imread('assets/image1.png')

# Flipping
flipped = cv2.flip(image, 1) # horizontal flip for other 0 or -1

# Rotation
rotated = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

# Cropping
x_start, y_tart = 30, 40
x_end, y_end = 130, 140

cropped = image[y_tart:y_end, x_start:x_end]

plt.figure(figsize=(10, 8))

plt.subplot(1,3,1)
plt.imshow(flipped)
plt.title('flip')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(rotated)
plt.title('rotation')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(cropped)
plt.title('cropping')
plt.axis('off')

```

```python
# Drawing lines and shapes 
image = image =cv2.imread('assets/image1.png')

# lines
line_image = image.copy()
start = (20,100)
end = (100,59)
line = cv2.line(line_image, start, end, (0,0,0), 5)

# rectangles
rect_image = image.copy()
pt1 = (20,100)
pt2 = (100,59)
rectangle = cv2.rectangle(rect_image, pt1, pt2, (0,0,0), 5)

# circle
circ_image = image.copy()
center = (150,150)
circle = cv2.circle(circ_image, center, 20, (0,0,0), 5)

# polygone
poly_image = image.copy()
pts = np.array([[100,100],
                [400,100],
                [250,400],
                [100,400]], np.int32)

pts = pts.reshape((-1,1,2))
polygone =cv2.polylines(poly_image, [pts], True,(0,0,0), 5)

# adding text
image_text = image.copy()
text = 'Life is good!!'
text_image = cv2.putText(image, text, (150,150),cv2.FONT_HERSHEY_COMPLEX, 1.3,(0,0,0), 5, cv2.LINE_AA)
plt.imshow(text_image)
plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(line)
plt.title('line')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(rectangle)
plt.title('rectangle')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(circle)
plt.title('circle')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(polygone)
plt.title('polygone')
plt.axis('off')

```

```python
# Affine and perspective transformations
image = cv2.imread("assets/image1.png")
rows, cols = image.shape[:2]

# Affine transformation
pts1 = np.float32([
    [50, 50],
    [200, 50],
    [50, 200]
])

pts2 = np.float32([
    [10, 100],
    [200, 50],
    [100, 250]
])

matrix = cv2.getAffineTransform(pts1, pts2)

affined = cv2.warpAffine(image, matrix, (cols, rows))

# Perspective transformation
pts1 = np.float32([
    [100, 150],
    [400, 120],
    [420, 400],
    [80, 420]
])

pts2 = np.float32([
    [0, 0],
    [300, 0],
    [300, 400],
    [0, 400]
])
matrix = cv2.getPerspectiveTransform(pts1, pts2)

perspected = cv2.warpPerspective(image, matrix, (cols, rows))


plt.figure(figsize=(10, 8))

plt.subplot(1,2,1)
plt.imshow(affined)
plt.title('affine transformation')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(perspected)
plt.title('perspective transformation')
plt.axis('off')

```

```python
# Blurring
image_bgr =cv2.imread('assets/image1.png')
image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
kernel_size =(11,11)

# average blur
blur = cv2.blur(image, kernel_size)

# gaussien blur 
std = 4
g_blur = cv2.GaussianBlur(image, kernel_size, std)

# median blur
m_blur = cv2.medianBlur(image, kernel_size[0])

plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(image)
plt.title('original')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(blur)
plt.title('average blur')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(g_blur)
plt.title('gaussien blur')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(m_blur)
plt.title('median blur')
plt.axis('off')
```
