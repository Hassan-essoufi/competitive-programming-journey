```python
# Three sum
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        if len(nums) == 3:
            if sum(nums) == 0:
                return [nums]
            else:
                return []
        l =[]
        i = 0
        nums = sorted(nums)
        while i <len(nums)-1:
            target = nums[i]
            j = i+1
            k = len(nums)-1
            while j < k:
                if nums[j] + nums[k] == -target:
                    h = sorted([nums[j], nums[k], target])
                    if h not in l:
                        l.append(h)
                    j = j+1
                    k = k-1
                elif nums[j] + nums[k] < -target:
                    j = j+1
                else :
                    k = k-1
            i = i+1
        return l
```

```python
# Is palindrome

class Solution(object):
    def isPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """
        if len(s) == 0:
            return True
        s = s.lower().strip()
        s = "".join([c for c in s if c.isalnum()])
        left = 0
        right = len(s)-1
        while left <= right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1
        return True

```

```python
# Container with most water

class Solution:
    def maxArea(self, heights: List[int]) -> int:
        if len(heights) == 2:
            return min(heights[0], heights[1])
        d = 0
        for i in range(len(heights)-1):
            for j in range(i+1, len(heights)):
                m  = min(heights[j], heights[i])
                d = max(d, (j-i)*m)

        return d
```

```python
# Rotate array

class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        k = k % len(nums)
        for i in range(len(nums)//2):
            nums[i] , nums[len(nums)-1-i] = nums[len(nums)-1-i], nums[i]
        for i in range(k//2):
            nums[i] , nums[k-1-i] = nums[k-1-i], nums[i]
        
        for i in range((len(nums)-k)//2):
            nums[i+k] , nums[len(nums)-1-i] = nums[len(nums)-1-i], nums[i+k]
        


```

```python
# container wtith most water
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        if len(height) == 2:
            return min(height)
        r = 0
        left = 0
        right = len(height)-1
        while left < right:
            m = min(height[right], height[left])
            r = max(r, m*(right-left))
            if height[right] > height[left]:
                left = left + 1
            else:
                right = right-1
        return r
```
