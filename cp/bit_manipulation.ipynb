```python
#Number of 1 Bits
from collections import Counter
class Solution:
    def hammingWeight(self, n: int) -> int:
        h = []
        while n >= 1:
            h.append(n%2)
            n = n//2
        counted = Counter(h)
        return counted[1]
```

```python
#Counting bits
from collections import Counter
class Solution:
    def countBits(self, n: int) -> List[int]:
        l = []
        for i in range(n+1):
            h = []
            while i >=1:
                h.append(i%2)
                i = i//2
            l.append(Counter(h)[1])
        return l
```
