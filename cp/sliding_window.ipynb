```python
# Maximum Average Subarray I
class Solution(object):
    def findMaxAverage(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: float
        """
        if len(nums) == 1:
            return float(nums[0])
        current = sum(nums[:k])
        max_sub = current
        for i in range(k,len(nums)):
            current = current-nums[i-k] +nums[i]
            max_sub = max(current, max_sub)
        return float(max_sub)/k
```

```python
# Remove duplicates
class Solution(object):
    def containsNearbyDuplicate(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: bool
        """
        if len(nums) == 1:
            return True if k == 0 else False
        window = set()
        
        for i in range(len(nums)):
            if nums[i] in window:
                return True
            if nums[i] not in window:
                window.add(nums[i])
            if len(window)> k:
                window.remove(nums[i-k]) 
        return False
```

```python
# Longest Harmonious Subsequence
from collections import Counter
class Solution(object):
    def findLHS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if len(nums) == 1:
            return 0
        frequences = Counter(nums)
        max_subsequence = 0
        for num in frequences:
            if num+1 in frequences:
                current = frequences[num] + frequences[num+1]
                max_subsequence = max(current, max_subsequence)

        return max_subsequence
```
